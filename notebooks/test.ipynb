```python
test_det = np.array([[1, 2], [3, 4]])
np.linalg.det(test_det)
```

```python
test_sum = np.array([[1, 2], [3, 4]])
np.sum(test_sum)
```

```python
test_prod = np.array([[1, 2], [3, 4]])
np.prod(test_prod)
```

```python
test_diag = np.array([[1, 2], [3, 4]])
np.diag(test_diag)
```

```python
test_symmetric = np.random.rand(2, 2)
print("previous:", test_symmetric)
test_symmetric[0][1] = test_symmetric[1][0]
print("current:", test_symmetric)
```
